```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
economic_file = "economic2019.csv"
economic_df = pd.read_csv(economic_file)
economic_df
```

```python
happy_file = "happy2019.csv"
happy_df = pd.read_csv(happy_file)
happy_df.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/felicidad2019"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
economic_df.to_sql(name='economic', con=engine, if_exists='append', index=True)
```

```python
happy_df.to_sql(name='happy', con=engine, if_exists='append', index=True)
```
